```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
import joblib
```

```python
clean_df = pd.read_csv(
    "../data/cleaned/online_retail_cleaned.csv",
    parse_dates=["InvoiceDate"]
)

clean_df.head()
```

```python
customer_product_matrix = pd.pivot_table(
    clean_df,
    index="CustomerID",
    columns="Description",
    values="Quantity",
    aggfunc="sum",
    fill_value=0
)

customer_product_matrix.head()
```

```python
customer_product_matrix = (customer_product_matrix > 0).astype(int)

customer_product_matrix.head()
```

```python
product_customer_matrix = customer_product_matrix.T

product_customer_matrix.head()
```

```python
similarity_matrix = cosine_similarity(product_customer_matrix)

similarity_df = pd.DataFrame(
    similarity_matrix,
    index=product_customer_matrix.index,
    columns=product_customer_matrix.index
)

similarity_df.head()
```

```python
def recommend_products(product_name, similarity_df, top_n=5):
    """
    Recommend products similar to the given product.
    """

    if product_name not in similarity_df.index:
        return ["Product not found."]

    recommendations = (
        similarity_df[product_name]
        .sort_values(ascending=False)
        .iloc[1 : top_n + 1]
    )

    return recommendations.index.tolist()
```

```python
clean_df["Description"].sort_values().unique()[:20]

recommend_products(
    "WHITE HANGING HEART T-LIGHT HOLDER",
    similarity_df
)
```

```python
joblib.dump(
    similarity_df,
    "../models/similarity_matrix.pkl"
)

print("Similarity matrix saved successfully!")
```

```python
product_names = similarity_df.index.tolist()

joblib.dump(
    product_names,
    "../models/product_names.pkl"
)

print("Product names saved successfully!")
```

| Step                    | Description                        |
| ----------------------- | ---------------------------------- |
| Customer–Product Matrix | Represents customer purchases      |
| Binary Encoding         | Indicates purchase or no purchase  |
| Matrix Transpose        | Converts to product-level analysis |
| Cosine Similarity       | Measures product similarity        |
| Recommendation Function | Returns the top 5 similar products |
