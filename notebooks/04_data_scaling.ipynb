```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
import joblib
```

```python
rfm = pd.read_csv("../data/processed/rfm_dataset.csv")

rfm.head()
```

```python
rfm_features = rfm[["Recency", "Frequency", "Monetary"]]

rfm_features.describe()
```

```python
scaler = StandardScaler()

scaled_features = scaler.fit_transform(rfm_features)
```

```python
scaled_rfm = pd.DataFrame(
    scaled_features,
    columns=rfm_features.columns
)

scaled_rfm.head()
```

```python
scaled_rfm.describe().round(2)
```

```python
joblib.dump(
    scaler,
    "../models/scaler.pkl"
)

print("Scaler saved successfully!")
```

```python
scaled_rfm.to_csv(
    "../data/processed/scaled_rfm.csv",
    index=False
)

print("Scaled dataset saved successfully!")
```
