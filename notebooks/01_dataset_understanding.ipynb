```python
# Create a copy of the original dataset
clean_df = df.copy()

print("Original Shape:", clean_df.shape)
```

```python
# Before cleaning
before_rows = clean_df.shape[0]

# Remove rows with missing CustomerID
clean_df = clean_df.dropna(subset=["CustomerID"])

# After cleaning
after_rows = clean_df.shape[0]

print(f"Rows Before: {before_rows}")
print(f"Rows After : {after_rows}")
print(f"Rows Removed: {before_rows - after_rows}")
```

```python
before_rows = clean_df.shape[0]

clean_df = clean_df.dropna(subset=["Description"])

after_rows = clean_df.shape[0]

print(f"Rows Before: {before_rows}")
print(f"Rows After : {after_rows}")
print(f"Rows Removed: {before_rows - after_rows}")
```

```python
before_rows = clean_df.shape[0]

clean_df = clean_df.drop_duplicates()

after_rows = clean_df.shape[0]

print(f"Rows Before: {before_rows}")
print(f"Rows After : {after_rows}")
print(f"Rows Removed: {before_rows - after_rows}")
```

```python
before_rows = clean_df.shape[0]

clean_df = clean_df[~clean_df["InvoiceNo"].str.startswith("C")]

after_rows = clean_df.shape[0]

print(f"Rows Before: {before_rows}")
print(f"Rows After : {after_rows}")
print(f"Rows Removed: {before_rows - after_rows}")
```

```python
before_rows = clean_df.shape[0]

clean_df = clean_df[clean_df["Quantity"] > 0]

after_rows = clean_df.shape[0]

print(f"Rows Before: {before_rows}")
print(f"Rows After : {after_rows}")
print(f"Rows Removed: {before_rows - after_rows}")
```

```python
before_rows = clean_df.shape[0]

clean_df = clean_df[clean_df["UnitPrice"] > 0]

after_rows = clean_df.shape[0]

print(f"Rows Before: {before_rows}")
print(f"Rows After : {after_rows}")
print(f"Rows Removed: {before_rows - after_rows}")
```

```python
print("Final Dataset Shape:", clean_df.shape)

print("\nRemaining Missing Values:")
print(clean_df.isnull().sum())

print("\nRemaining Duplicate Rows:")
print(clean_df.duplicated().sum())

print("\nInvalid Quantity Rows:")
print((clean_df["Quantity"] <= 0).sum())

print("\nInvalid Unit Price Rows:")
print((clean_df["UnitPrice"] <= 0).sum())

print("\nCancelled Orders:")
print(clean_df["InvoiceNo"].str.startswith("C").sum())
```

```python
print("=" * 60)
print("FINAL DATA VALIDATION REPORT")
print("=" * 60)

print("\nDataset Shape")
print(clean_df.shape)

print("\nMissing Values")
print(clean_df.isnull().sum())

print("\nDuplicate Rows")
print(clean_df.duplicated().sum())

print("\nNegative Quantity")
print((clean_df["Quantity"] <= 0).sum())

print("\nNegative Unit Price")
print((clean_df["UnitPrice"] <= 0).sum())

print("\nCancelled Orders")
print(clean_df["InvoiceNo"].str.startswith("C").sum())
```
